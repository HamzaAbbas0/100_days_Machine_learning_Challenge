# OneHotEncoding using Sklearn

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_train = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=True)
```

```python
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first',sparse=False,dtype =np.int32)
# drop fist means ka one hot encode time 1st column remove ho ga tka multi_colrelatily na ho ..
# sparse False means hai ka numpy array ma convert ho jy ga
```

```python
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
```

```python
X_train_new.shape
```

```python
X_test_new = ohe.transform(X_test[['fuel','owner']])
```

# 4. One Hot Encoding With Top Catagories

```python
df['brand'].value_counts()
```

```python
brand_counts = df['brand'].value_counts()

threshold = 100
```

```python
repl = brand_counts[brand_counts <= threshold].index
```

```python
repl.shape
```

```python
pd.get_dummies(df['brand'].replace(repl,'uncommon')).sample(5)
```
