```python
# Count the number of people with age equal to 20
count_20_age = df[df['Age'] == 20].shape[0]

# Create a bar plot
plt.figure(figsize=(6, 4))
ax =sns.barplot(x=['Age 20'], y=[count_20_age])
# Display the count values on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')
plt.title('Number of People with Age 20')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax=sns.barplot(data=df1, x=city_counts.index, order=city_counts.index, y=city_counts.values)
# Display the count values on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.xlabel('City')
plt.ylabel('Number of Matches')
plt.title('Top 10 Cities with Most Matches Played')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the Titanic dataset
titanic_data = pd.read_csv('train.csv')  # Replace with the actual path

# Create a figure with subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Plot 1: Survival Rate by Gender
sns.barplot(x='Sex', y='Survived', data=titanic_data, ax=axes[0, 0])
axes[0, 0].set_title('Survival Rate by Gender')

# Plot 2: Survival Rate by Passenger Class
sns.barplot(x='Pclass', y='Survived', data=titanic_data, ax=axes[0, 1])
axes[0, 1].set_title('Survival Rate by Passenger Class')

# Plot 3: Age Distribution
sns.histplot(data=titanic_data, x='Age', bins=20, ax=axes[0, 2])
axes[0, 2].set_title('Age Distribution')

# Plot 4: Fare Distribution
sns.histplot(data=titanic_data, x='Fare', bins=20, ax=axes[1, 0])
axes[1, 0].set_title('Fare Distribution')

# Plot 5: Embarkation Port Distribution
sns.countplot(x='Embarked', data=titanic_data, ax=axes[1, 1])
axes[1, 1].set_title('Embarkation Port Distribution')

# Plot 6: Survival Rate by Embarkation Port
sns.barplot(x='Embarked', y='Survived', data=titanic_data, ax=axes[1, 2])
axes[1, 2].set_title('Survival Rate by Embarkation Port')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
